## **Stemming**

```python
## classification problem
## comments on the product is a positive or negative review
## reviews --> eating, eat, eaten (same kind of  word --> word stem of eat) --> eat
words = ["eating", "eats", "writing", "writes", "eaten", "programming", "programs", "history", "finally", "finalize"]
```

### **PorterStemmer**

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+"--->"+stemming.stem(word))
```

```python
stemming.stem("Congratulation")
```

```python
stemming.stem("Sitting")
```

### **RegexpStemmer Class**

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

### **Snowball Stemmer**

```python
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(word+"---->"+snowballstemmer.stem(word))
```

```python
stemming.stem('fairly'), stemming.stem('sportingly'), stemming.stem('goes')
```

```python
snowballstemmer.stem('fairly'), snowballstemmer.stem('sportingly'), snowballstemmer.stem('goes')
```
